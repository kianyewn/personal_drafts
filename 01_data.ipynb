```python
import joblib
    
class DataProcessor:
    def __init__(self, config: Dict, **kwargs):
        self.config = config
        self.numerical_encoder = self.kwargs.get('numerical_encoder', self.load_numerical_encoder(self.config['numerical_encoder']))
        self.categorical_encoder = self.kwargs.get('categorical_encoder', self.load_categorical_encoder(self.config['categorical_encoder']))

    def load_numerical_encoder(self, path):
        numerical_encoder = NumericalEncoder(self.config['numerical_features'])
        return numerical_encoder

    def load_categorical_encoder(self, path):
        categorical_encoder = CategoricalEncoder(self.config['categorical_features'])
        return categorical_encoder

    def fit_numerical_encoder(self, df):
        self.numerical_encoder.fit(df[self.config['numerical_features']])
        df[self.config['numerical_features']] = self.numerical_encoder.transform(df[self.config['numerical_features']])
        return df

    def fit_categorical_encoder(self, df):
        self.categorical_encoder.fit(df[self.config['categorical_features']])
        df[self.config['categorical_features']] = self.categorical_encoder.transform(df[self.config['categorical_features']])
        return df
     
    def fit(self, train_df):
        self.fit_numerical_encoder(train_df)
        self.fit_categorical_encoder(train_df)
        return 
    
    def process(self, df):
        df = self.numerical_encoder.transform(df)
        df = self.categorical_encoder.transform(df)
        return df

    def transform(self, df):
        df = self.numerical_encoder.transform(df)
        df = self.categorical_encoder.transform(df)
        return df 
    
    def save(self, path: str) -> None:
        """Save entire processor state in one file"""
        state = {
            'config': self.config,
            'is_initialized': self.is_initialized,
            'numerical_encoder': self.numerical_encoder,
            'categorical_encoder': self.categorical_encoder,
            'version': '1.0',
            'timestamp': datetime.now().isoformat()
        }
        
        joblib.dump(state, path)
        logger.info(f"Saved complete processor state to: {path}")
    
    def load(self, path: str) -> 'DataProcessor':
        """Load complete processor state from one file"""
        if not Path(path).exists():
            raise FileNotFoundError(f"Processor file not found: {path}")
        
        try:
            state = joblib.load(path)
            
            # Restore state
            self.config = state['config']
            self.is_initialized = state['is_initialized']
            self.numerical_encoder = state['numerical_encoder']
            self.categorical_encoder = state['categorical_encoder']
            
            logger.info(f"Loaded complete processor state from: {path}")
            return self
            
        except Exception as e:
            raise ValueError(f"Failed to load processor from {path}: {e}")
    
    @classmethod
    def from_file(cls, path: str) -> 'DataProcessor':
        """Create processor instance from saved file"""
        if not Path(path).exists():
            raise FileNotFoundError(f"Processor file not found: {path}")
        
        try:
            state = joblib.load(path)
            processor = cls(state['config'])
            
            # Restore state
            processor.is_initialized = state['is_initialized']
            processor.numerical_encoder = state['numerical_encoder']
            processor.categorical_encoder = state['categorical_encoder']
            
            return processor
            
        except Exception as e:
            raise ValueError(f"Failed to create processor from {path}: {e}")

dp = DataProcessor(config, numerical_encoder=NumericalEncoder), categorical_encoder=CategoricalEncoder))
dp.fit(train)
dp.save('data_processor.pkl')
dp.load('data_processor.pkl')
train = dp.transform(train)
valid = dp.transform(valid)
test = dp.transform(test)
```

```python
categorical_features = ['gender', 'genre']

class Model(nn.Module):
    def __init__(self):
        super().__init__()
        self.genre_hist_emb  = nn.Embedding(10, 10)
        self.current_item_emb = nn.Embedding(10, 10)
        self.item_hist_emb = nn.Embedding(10, 10)
        
    def forward(self, x):
        current_item_emb = self.current_item_emb(x['current_item'].type(torch.long))
        history_emb = self.genre_hist_emb(torch.stack(x['sequence_info']['genre']).T.type(torch.long))
        return current_item_emb, history_emb
    
    @classmethod
    def get_pretrained_model(cls, model_path, map_location=None):
        model_data = torch.load(model_path, map_location=map_location)
        model.load_state_dict(model_data)
        model.eval()
        for p in model.parameters():
            p.requires_grad_(False)
        return model

    def save_model(self, model_path):
        torch.save(self.state_dict(), model_path)
        logger.info(f'Model {self.__class__.__name__} saved to {model_path}')
        return

model = Model()
model(batch)[0].shape # B, T, D
```
